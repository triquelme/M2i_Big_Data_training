```python
from selenium import webdriver


#Connexion chrome
webpage = "https://www.imdb.com/search/title/" # edit me
driver = webdriver.Chrome("/Users/thomas/Documents/poe_bigdata/s10_mini_project/chromedriver")
driver.get(webpage)


#Remplissage des champs
min_date = "2015-01-01" 
max_date = "2020-01-01"

sbox_min = driver.find_element_by_name("release_date-min")
sbox_min.send_keys(min_date)

sbox_max = driver.find_element_by_name("release_date-max")
sbox_max.send_keys(max_date)

submit = driver.find_element_by_class_name("primary")
submit.click()


#Collecte lien vers la page du film
list_films = driver.find_elements_by_class_name("lister-item-header")

film_links = []
for film in list_films:    
    i = list_films.index(film)+1
    
    film_link_bloc = driver.find_element_by_xpath('//*[@id="main"]/div/div[3]/div/div['+str(i)+']/div[3]/h3/a')
    film_link = film_link_bloc.get_attribute('href')
    film_links.append(film_link)
    
    
liste_serie_pour_json = []
liste_img = []
    
#Collecte des donées par film    
for link in film_links:
    driver.get(link)
    
    print(link)
    
    title = driver.find_element_by_xpath('//*[@class="title_wrapper"]/h1').text
    print('Title: ', title)
    
    ratingValue = driver.find_element_by_class_name("ratingValue").text
    print('ratingValue: ', ratingValue)
    
    storyLine = driver.find_element_by_xpath('//*[@class="inline canwrap"]/p').text
    print('story Line: ', storyLine)
    
    country = driver.find_element_by_xpath('//*[@id="titleDetails"]/div[1]').text
    print(country)
    
    language = driver.find_element_by_xpath('//*[@id="titleDetails"]/div[2]').text
    print(language)
    
    releaseDate = driver.find_element_by_xpath('//*[@id="titleDetails"]/div[3]').text
    print(releaseDate)
    
    runTime = driver.find_element_by_xpath('//*[@id="titleDetails"]/div[8]').text
    print(runTime)

    try:
        img = driver.find_element_by_xpath('//*[@id="title-overview-widget"]/div[1]/div[4]/div[1]/a/img')
    except:
        try:
            img = driver.find_element_by_xpath('//*[@id="title-overview-widget"]/div[1]/div[3]/div[1]/a/img')          
        except:    
            print("error")
            pass
        
    print('-----------------------------------')
    
    dict_serie = {"title": title}
    liste_serie_pour_json.append(dict_serie)
    liste_img.append(img)
    # Rq: on ne récupère que le lien url vers le poster du film, pas l'image en elle même
    # pour récupérer l'image:
    #urllib.request.urlretrieve(img, "C://Users//admin//Desktop//"+title+"_poster.jpg")
    #client.upload("C://Users//admin//Desktop//"+title+"_poster.jpg", '/data_img')
    #os.delete("C://Users//admin//Desktop//poster_serie_1.jpg")    
    # Pseudo code à vérifier
    
```

```python
from hdfs import InsecureClient   
    

def connect_to_hdfs(host):
    try:
        client = InsecureClient(host)   
        print('Connected to hdfs.')
    except:
        print('ERROR connecting')
    
    return client

HOST_HDFS = 'http://localhost:9870'  
#connect to hdfs
client =  connect_to_hdfs(HOST_HDFS)     
    

#Create directory in HDFS to save jsons
hdfs_path = '/imdb_data'
client.makedirs(hdfs_path)




def write_to_hdfs(client, liste_serie_pour_json):
    from json import dump
    # As a context manager:
    with client.write(hdfs_path+'/imdb_serie.jsonl', encoding='utf-8') as writer:
        dump(liste_serie_pour_json, writer)

        
#write to hdfs json file                              
write_to_hdfs(client, liste_serie_pour_json) 
    
```
