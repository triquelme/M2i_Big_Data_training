```python
import kafka
from kafka import KafkaProducer
from json import dumps
import xml.etree.ElementTree as ET

if __name__ == "__main__":

    tree = ET.parse('fruits.xml')

    fruits_liste = []
    root = tree.getroot()

    for child in root:
        fruits_liste.append({'nom': child.find('nom').text,'couleur': child.find('couleur').text, 'taille': child.find('taille').text })

    producer = KafkaProducer(bootstrap_servers=['localhost:9092'], 
                                 value_serializer=lambda x: dumps(x).encode('utf-8'))
    
    producer.send('topic-cr', value=fruits_liste)
```

```python
import pandas as pd

df = pd.DataFrame(fruits_liste)

df
```
